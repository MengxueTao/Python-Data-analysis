### ndarray与Python原生list效率对比

```python
import random
import time
```

```python
# 生成一个大数组
python_list=[]

for i in range(100000000):
    python_list.append(random.random())
```

```python
ndarray_list=np.array(python_list)
ndarray_list
```

```python
len(ndarray_list)
```

# 原生pythonlist求和

```python
t1=time.time()
a=sum(python_list)
t2=time.time()
d1=t2-t1

t3=time.time()
b=np.sum(ndarray_list)
t4=time.time()
d2=t4-t3
```
